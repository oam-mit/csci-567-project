```python
df["MasVnrType"] = df["MasVnrType"].fillna("None")
```

```python
df.columns[df.isna().any()].tolist()
```

```python
def replace_with_mean(df, column):
    mean = df[column].mean()
    df[column] = df[column].fillna(mean)
    return df
```

```python
df = replace_with_mean(df, column="LotFrontage")
```

```python
df = replace_with_mean(df, column="MasVnrArea")
```
