```python
def replace_with_mean(df, column, to_int = False):
    mean = df[column].mean()
    if to_int:
        df[column] = df[column].fillna(int(mean))
    else:
        df[column] = df[column].fillna(mean)
    return df
```

```python
utilites_map = {"AllPub" : 0, "NoSewr": 1, "NoSeWa":2, "ELO":3}
condition_2_map = {index: key for key, index in enumerate(df["Condition1"].astype("category").cat.categories)}
exterior_list = {index: key for key, index in enumerate(["AsbShng", "AsphShn", "BrkComm", "BrkFace", "CBlock", "CemntBd", "HdBoard", "ImStucc", "MetalSd", "Other", "Plywood", "PreCast", "Stone", "Stucco", "VinylSd", "Wd Sdng", "WdShing"])}
masvnrtype_map = {index:key for key, index in enumerate(["BrkCmn", "BrkFace", "CBlock", "None", "Stone"])}
exterior_quality_map = kitchen_qual_map = {index:key for key, index in enumerate(["Ex", "Gd", "TA", "Fa", "Po"])}
bsmt_quality_map = bsmt_cond_map  = {index:key for key, index in enumerate(["Ex", "Gd", "TA", "Fa", "Po", "NA"])}
functional_map = {index:key for key, index in enumerate(["Typ", "Min1", "Min2", "Mod", "Maj1", "Maj2", "Sev", "Sal"])}
pool_qc_map = {index:key for key, index in enumerate(["Ex", "Gd", "TA", "Fa", "NA"])}
misc_feature_map = {index:key for key, index in enumerate(["Elev", "Gar2", "Othr", "Shed", "TenC", "NA"])}
sales_type_feature_map = {index:key for key, index in enumerate(["WD", "CWD", "VWD", "New", "COD", "Con", "COD", "Con", "ConLw" ,"ConLI", "ConLD", "Oth"])}
```

```python
class Mapping():
    def __init__(self, name, mapping):
        self.name = name
        self.mapping = mapping
        
```

```python
from typing import List


def get_proprocessed_data():
    mapping:List[Mapping] = []


    category_list = ["Street", "Alley", "LotShape", "LandContour", "LotConfig", "LandSlope", "BldgType", "HouseStyle", "RoofStyle", "RoofMatl", "ExterCond", "BsmtExposure", "BsmtFinType1", "BsmtFinType2", "Heating", "HeatingQC", "CentralAir", "Electrical", "FireplaceQu", "GarageType", "GarageFinish", "GarageQual", "GarageCond", "Fence", "SaleCondition","Neighborhood", "Foundation", "PavedDrive"]

    df["MSSubClass"] = df["MSSubClass"].replace(MS_SubclassMapping)
    mapping.append(Mapping("MSSubClass", MS_SubclassMapping))

    df["MSZoning"] = df["MSZoning"].replace(MSZoning_Mapping)
    mapping.append(Mapping("MSZoning", MSZoning_Mapping))

    df["Utilities"] = df["Utilities"].replace(utilites_map)
    mapping.append(Mapping("Utilities", utilites_map))

    df["Condition1"] = df["Condition1"].replace(condition_2_map)
    mapping.append(Mapping("Condition1", condition_2_map))

    df["Condition2"] = df["Condition2"].replace(condition_2_map)
    mapping.append(Mapping("Condition2", condition_2_map))


    df["Exterior1st"] = df["Exterior1st"].replace(exterior_list)
    mapping.append(Mapping("Exterior1st", exterior_list))

    df["Exterior2nd"] = df["Exterior2nd"].replace(exterior_list)
    mapping.append(Mapping("Exterior2nd", exterior_list))

    df["MasVnrType"] = df["MasVnrType"].replace(masvnrtype_map)
    mapping.append(Mapping("MasVnrType", masvnrtype_map))
    
    df["ExterQual"] = df["ExterQual"].replace(exterior_quality_map)
    mapping.append(Mapping("ExterQual", exterior_quality_map))

    df["BsmtQual"] = df["BsmtQual"].replace(bsmt_quality_map)
    mapping.append(Mapping("BsmtQual", bsmt_quality_map))

    df["BsmtCond"] = df["BsmtCond"].replace(bsmt_cond_map)
    mapping.append(Mapping("BsmtCond", bsmt_cond_map))

    df["KitchenQual"] = df["KitchenQual"].replace(kitchen_qual_map)
    mapping.append(Mapping("KitchenQual", kitchen_qual_map))

    df["Functional"] = df["Functional"].replace(functional_map)
    mapping.append(Mapping("Functional", functional_map))

    df["PoolQC"] = df["PoolQC"].replace(pool_qc_map)
    mapping.append(Mapping("PoolQC", pool_qc_map))

    df["MiscFeature"] = df["MiscFeature"].replace(misc_feature_map)
    mapping.append(Mapping("MiscFeature", misc_feature_map))

    df["SaleType"] = df["SaleType"].replace(sales_type_feature_map)
    mapping.append(Mapping("SaleType", sales_type_feature_map))


    for category in category_list:
        df[category] = df[category].astype("category")
        category_mappings = {category_label: category_index for category_index, category_label in enumerate(df[category].cat.categories)}
        mapping.append(Mapping(category, category_mappings))
        df[category] = df[category].cat.codes
    

    return mapping
```

```python
mappings = get_proprocessed_data()
```
