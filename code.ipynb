# Data Preprocessing

```python
df["Alley"] = df["Alley"].fillna("NA")
```

```python
na_columns = ["Alley", "BsmtQual", "BsmtCond", "BsmtExposure", "BsmtFinType1", "BsmtFinType2", "FireplaceQu", "GarageType", "GarageFinish", "GarageQual", "GarageCond", "PoolQC", "Fence", "MiscFeature"]
```

```python
df["MasVnrType"] = df["MasVnrType"].fillna("None")
```

```python
df.columns[df.isna().any()].tolist()
```

```python
def replace_with_mean(df, column):
    mean = df[column].mean()
    df[column] = df[column].fillna(mean)
    return df
```

```python
df = replace_with_mean(df, column="LotFrontage")
```

```python
df = replace_with_mean(df, column="MasVnrArea")
```

```python
df.columns[df.isna().any()].tolist()
```

```python
df["Electrical"] = df["Electrical"].fillna("SBrkr")
```

```python
df["GarageYrBlt"] = df["GarageYrBlt"].fillna(2005)
```

```python
MS_SubclassMapping = {20:0, 30:1, 40:2, 45:3, 50:4, 60:5, 70: 6, 75: 7, 80: 8, 85:9, 90: 10, 120: 11, 150: 12, 160: 13, 180: 14, 190: 15}
```

```python
MSZoning_Mapping = {"A":0, "C":1, "FV":2, "I":3, "RH":4, "RL":5, "RP":6, "RM":7}
```

```python
utilites_map = {"AllPub" : 0, "NoSewr": 1, "NoSeWa":2, "ELO":3}
condition_2_map = {key: index for key, index in enumerate(df["Condition1"].astype("category").cat.categories)}
exterior_list = {key: index for key, index in enumerate(["AsbShng", "AsphShn", "BrkComm", "BrkFace", "CBlock", "CemntBd", "HdBoard", "ImStucc", "MetalSd", "Other", "Plywood", "PreCast", "Stone", "Stucco", "VinylSd", "Wd Sdng", "WdShing"])}
masvnrtype_map = {index:key for key, index in enumerate(["BrkCmn", "BrkFace", "CBlock", "None", "Stone"])}
exterior_quality_map = kitchen_qual_map = {index:key for key, index in enumerate(["Ex", "Gd", "TA", "Fa", "Po"])}
bsmt_quality_map = bsmt_cond_map  = {index:key for key, index in enumerate(["Ex", "Gd", "TA", "Fa", "Po", "NA"])}
functional_map = {index:key for key, index in enumerate(["Typ", "Min1", "Min2", "Mod", "Maj1", "Maj2", "Sev", "Sal"])}
pool_qc_map = {index:key for key, index in enumerate(["Ex", "Gd", "TA", "Fa", "NA"])}
misc_feature_map = {index:key for key, index in enumerate(["Elev", "Gar2", "Othr", "Shed", "TenC", "NA"])}
sales_type_feature_map = {index:key for key, index in enumerate(["WD", "CWD", "VWD", "New", "COD", "Con", "COD", "Con", "ConLw" ,"ConLI", "ConLD", "Oth"])}
```

```python
def get_proprocessed_data():

    category_list = ["Street", "Alley", "LotShape", "LandContour", "LotConfig", "LandSlope", "BldgType", "HouseStyle", "RoofStyle", "RoofMatl", "ExterCond", "BsmtExposure", "BsmtFinType1", "BsmtFinType2", "Heating", "HeatingQC", "CentralAir", "Electrical", "FireplaceQu", "GarageType", "GarageFinish", "GarageQual", "GarageCond", "Fence", "SaleCondition"]
    df["MSSubClass"] = df["MSSubClass"].replace(MS_SubclassMapping)
    df["MSZoning"] = df["MSZoning"].replace(MSZoning_Mapping)
    df["Utilities"] = df["Utilities"].replace(utilites_map)
    df["Condition1"] = df["Condition1"].replace(condition_2_map)
    df["Condition2"] = df["Condition2"].replace(condition_2_map)
    df["Exterior1st"] = df["Exterior1st"].replace(exterior_list)
    df["Exterior2nd"] = df["Exterior2nd"].replace(exterior_list)
    df["MasVnrType"] = df["MasVnrType"].replace(masvnrtype_map)
    df["ExterQual"] = df["ExterQual"].replace(exterior_quality_map)
    df["BsmtQual"] = df["BsmtQual"].replace(bsmt_quality_map)
    df["BsmtCond"] = df["BsmtCond"].replace(bsmt_cond_map)
    df["KitchenQual"] = df["KitchenQual"].replace(kitchen_qual_map)
    df["Functional"] = df["Functional"].replace(functional_map)
    df["PoolQC"] = df["PoolQC"].replace(pool_qc_map)
    df["MiscFeature"] = df["MiscFeature"].replace(misc_feature_map)
    df["SaleType"] = df["SaleType"].replace(sales_type_feature_map)

    for category in category_list:
        df[category] = df[category].astype("category")
        df[category] = df[category].cat.codes
```
